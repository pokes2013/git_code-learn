```python
import openpyxl
from openpyxl.styles import Alignment

# 创建一个新的Workbook对象
wb = openpyxl.Workbook()

# 获取默认的工作表
sheet = wb.active

# 三个需要写入的列表
list0 = ['番号','片名','发行时间']
list1 = ['DLDSS-241', '巨乳講師がレッスン中に拒みきれず身体を許したら性欲覚醒密室汗だく性交 美乃すずめ', '2023-12-07']
list2 = ['Data1', 'Data2', 'Data3']
list3 = ['Value1', 'Value2', 'Value3']


# 写入多个单元格（追加模式）
sheet.append(list0)
sheet.append(list1)
sheet.append(list2)
sheet.append(list3)


# 创建一个Alignment对象，设置文本水平和垂直居中
align = Alignment(horizontal='left', vertical='center')     #horizontal 水平对齐，vertical，上下对齐
# 遍历所有单元格，应用对齐设置
for row in sheet.iter_rows():
    for cell in row:
        cell.alignment = align



# 保存工作簿到文件
wb.save('pokes3.xlsx')
```
