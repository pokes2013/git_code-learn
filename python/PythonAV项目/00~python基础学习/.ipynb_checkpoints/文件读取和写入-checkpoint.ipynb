# 文件读取

```python
file=open(r'D:\code\python\03.文本读写\a.txt','r',encoding='UTF-8')
print(file.readlines())  # readlines() 读取全部内容并且放到列表中
file.close()

file=open(r'D:\code\python\03.文本读写\a.txt','r',encoding='UTF-8')
print(file.readline())  # readlines() 读取1行
file.close()
```

```python
file=open(r'D:\code\python\03.文本读写\b.txt','w',encoding='UTF-8')
file.write('helloworld')
file.close()
```

```python
with open("D:/code/python/04.文件操作/ceshi.txt", "r", encoding="UTF-8") as f:
    for line in f:
        print(line)
```
