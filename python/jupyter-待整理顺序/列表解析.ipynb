# 如何在列表，字典，集合中根据条件筛选数据   

## 列表解析 

data= [3,9,6,-1,10,20,-4,-5,-9] 过滤掉负数？

```python
data = [3, 9, 6, -1, 10, 20, -4, -5, -9]
result = [x for x in data if x >= 0]
print(result)
```

## 字典解析

下面字典中过滤出大于80的学生？  

```python
mydic = {'stu01': 65, 'stu02': 62, 'stu03': 95, 'stu04': 45, 'stu05': 85}
gg = {k: v for k, v in mydic.items() if v >= 80}
print(gg)
```
