前言

## nvm是什么？

nvm（node.js version management），是一个nodejs的版本管理工具。
nvm和n都是node.js版本管理工具，为了解决node.js 各种版本存在不兼容现象 可以通过它可以安装和切换不同版本的node.js。
可同时在一个环境中安装多个node.js版本（和配套的npm）

## nvm安装

### 下载nvm包

https://github.com/coreybutler/nvm-windows

### 安装nvm-setup.exe

按照界面操作，设置程序安装位置，和设置node包位置. 下一步 完成.

### 更改源

方法1：打开nvm安装包下的settings.txt文件，在末尾添加以下代码

````python
node_mirror: https://registry.npmmirror.com/node/
npm_mirror: https://registry.npmmirror.com/npm/
````

方法2：命令行添加

````python
nvm node_mirror https://registry.npmmirror.com/node/
nvm npm_mirror https://registry.npmmirror.com/npm/
````

### 安装版本

````python
C:\Users\Anita>nvm install v10.21.0
Downloading node.js version 10.21.0 (64-bit)...
Complete
Creating d:\APP\nvm\temp

Downloading npm version 6.14.4... Complete
Installing npm v6.14.4...

Installation complete. If you want to use this version, type

nvm use 10.21.0
````

````python
C:\Users\Anita>nvm install 20.16.0
Downloading node.js version 20.16.0 (64-bit)...
Extracting node and npm...
Complete
npm v10.8.1 installed successfully.
````

### 切换版本

````python
nvm use 20.16.0
nvm use 10.24.1
````
