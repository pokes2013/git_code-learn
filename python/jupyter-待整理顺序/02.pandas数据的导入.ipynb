# 数据导入

## 读取数据

```python
import pandas as pd
```

```python
df1 = pd.read_excel('测试数据.xlsx')
df1.head()
```

## 导入Excel的指定Sheet

```python
#第一种方式:sheetname工作表名称直接传输工作表名
df1 = pd.read_excel("测试数据.xlsx",sheet_name="sheet1")
df1.head()
```

```python
#第二种方式:sheet_rame 工作表名称传输索引号 索引号从0开始,0表示第一个工作表
df1 = pd.read_excel("测试数据.xlsx",sheet_name=0)
df1.head()
```

### 设置第1行为列索引

```python
df1 = pd.read_excel("测试数据.xlsx",sheet_name="sheet1",header=0)
df1.head()
```

### 通过指定的列索导入列

```python
# 导入第1列
df1 = pd.read_excel("测试数据.xlsx",sheet_name="sheet1",header=0,usecols=[0])
df1.head()
```

```python
# 导入第1列和第4列
df1 = pd.read_excel("测试数据.xlsx",sheet_name="sheet1",header=0,usecols=[0,3])
df1.head()
```

```python
# 通过指定的列名来导入指定列
df1 = pd.read_excel("测试数据.xlsx",sheet_name="sheet1",header=0,usecols=["单据号","客户"])
df1.head()
```
