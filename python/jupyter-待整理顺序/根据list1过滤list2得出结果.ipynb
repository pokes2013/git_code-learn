# 列表的过滤进阶版：根据list1过滤list2得出结果

根据list1过滤list2得出结果

```python
# coding=utf-8
# 声明两个列表变量
list1 = ['Python', 'PHP', 'Java', 'Bash']
list2 = ['JavaScript是客户端脚本语言',
         'PHP是服务器端脚本语言',
         'Java是一种编程语言',
         'Kotlin是一种静态编程语言']

# 根据第一个列表过滤第二个列表
filter_data = [x for x in list2 if all(y not in x for y in list1)]
print("过滤后的第二个列表的内容:", filter_data)
```
