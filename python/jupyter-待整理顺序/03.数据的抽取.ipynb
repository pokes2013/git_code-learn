# 数据抽取

## 数据生成

```python
import pandas as pd
```

```python
data = [
    [109,113,121],
    [108,124,141],
    [112,125,135],
    [132,113,126]
]

name = ['刘备','关羽','张飞','赵云']
columns = ['骑马','射箭','摔跤']
df = pd.DataFrame(data=data,index=name,columns=columns)
df
```

## 使用loc方法抽取

```python
df.loc['关羽']
```

```python
df.loc[['关羽','张飞']]
```

## 使用iloc方法抽取

### 使用索引号抽取

```python
df.iloc[1]
```

```python
df.iloc[[1,2]]
```

### 抽取指定行数据

```python
# 第1行到“关羽”
df.loc['刘备':'张飞']
```

```python
# 第1行到第4行
df.iloc[0:4]
```

```python
#第2行到最后1行
df.iloc[2:]
```

### 抽取指定列的数据

```python
#直接使用列名
df['射箭']
```

```python
df[['骑马','射箭']]
```

```python
# 抽取第1列和第2列
df.iloc[:,[0,1]]
```

```python
#抽取从“骑马”开始到最后1列
df.loc[:,'骑马':]
```
